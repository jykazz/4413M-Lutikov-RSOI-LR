```python
# xmlrpc_client.ipynb

import xmlrpc.client as rpc
import pickle

server_proxy = rpc.ServerProxy("http://localhost:8008")

print('Ping:', server_proxy.ping())

print('Server datetime:', server_proxy.now())

print('View, type, value:', server_proxy.exec('type', 2))
print('View, type, value:', server_proxy.exec('type', 2.))
print('View, type, value:', server_proxy.exec('type', 'My string'))
print('View, type, value:', server_proxy.exec('type', "My string"))
print('View, type, value:', server_proxy.exec('type', [1, 2, 3]))
print('View, type, value:', server_proxy.exec('type', ["one", "two", "three"]))
print('View, type, value:', server_proxy.exec('type', (1, 2, "3")))

print('Sum 2 + 3 :', server_proxy.sum(2, 3))
print('Pow 2^3: ', server_proxy.pow(2, 3))
```

```python
#     
def inv_color(img_arr_in):
    pimg = pickle.dumps(img_arr_in)
    img_bin = rpc.Binary(pimg)
    
    img_bin2 = server_proxy.color_inversion(img_bin)
    
    img_arr_out = pickle.loads(img_bin2.data)
    return img_arr_out

img_arr_inv = inv_color(img_arr)

#    
imshow(img_arr_inv)
```

```python
colored_img = '../../resources/lr1/Jellyfish.jpg'
monochrome_img = '../../resources/lr1/11.bmp'

def show_modified_image(path, modify_func, param=None):
    img = Image.open(path)
    img_arr = array(img)
    pimg = pickle.dumps(img_arr)
    img_bin = rpc.Binary(pimg)
    if param is None:
        img_bin_out = modify_func(img_bin)
    else:
        img_bin_out = modify_func(img_bin, param)
    img_arr_out = pickle.loads(img_bin_out.data) 
    gray()
    imshow(img_arr_out)

#  5.    send_back_inversion       ,
# ..      (M, N, 3)  (M, N, 1).

show_modified_image(colored_img, server_proxy.color_inversion)
```

```python
show_modified_image(monochrome_img, server_proxy.color_inversion)
```

```python
#  6.       1-255     

show_modified_image(colored_img, server_proxy.binarize_by_threshold, 100)
```

```python
show_modified_image(colored_img, server_proxy.binarize_by_threshold, 10)
```

```python
show_modified_image(monochrome_img, server_proxy.binarize_by_threshold, 100)
```

```python
#  7.          

show_modified_image(colored_img, server_proxy.mirror_vertically)
```

```python
show_modified_image(monochrome_img, server_proxy.mirror_vertically)
```
